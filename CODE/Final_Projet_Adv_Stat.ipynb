```python
row, col = D.shape

# The list of variables for professionnal and well-being questions (EWB)
list_of_flourishing_pro = ['MHCB1', 'MHCB2', 'MHCB3']

# The list of variables for personnal and well-being questions (EWB)
list_of_flourishing_home =  ['MHCC1', 'MHCC2', 'MHCC3']

# The list of variables for professionnal and (SWB & PWB)
list_of_positive_functioning_questions_pro = ['MHCB4', 'MHCB5', 'MHCB6', 'MHCB7', 'MHCB8', 'MHCB9', 'MHCB10', 'MHCB11', 'MHCB12', 'MHCB13', 'MHCB14']

# The list of variables for personnal and (SWB & PWB)
list_of_positive_functioning_questions_perso = ['MHCC4', 'MHCC5', 'MHCC6', 'MHCC7', 'MHCC8', 'MHCC9', 'MHCC10', 'MHCC13','MHCC14']

# For PRO : List that verify according to the sum of validate value if well-being questions is respected
list_of_satisfy_fourishing_pro = [0] * row;

# For HOME : List that verify according to the sum of validate value if well-being questions is respected
list_of_satisfy_fourishing_home = [0] * row;

# For PRO : List that verify according to the sum of validate value if languishing status is respected
list_of_satisfy_languishing_pro = [0] * row;

# For HOME : List that verify according to the sum of validate value if languishing status is respected
list_of_satisfy_languishing_home = [0] * row;

# For PRO : List that verify according to the sum of validate value if moderate status is respected
list_of_satisfy_moderate_pro = [0] * row;

# For HOME : List that verify according to the sum of validate value if moderate status is respected
list_of_satisfy_moderate_home = [0] * row;

# For PRO : List that verify according to the sum of validate value if positive functioning questions is respected
list_of_satisfy_functioning_questions_pro = [0] * row;

# For HOME : List that verify according to the sum of validate value if positive functioning questions is respected
list_of_satisfy_functioning_questions_home = [0] * row;

# Final column professionnal with the (SWB & PWB) & (EWB) restrictions
final_list_flourishing_pro = [0] * row;

# Final column personnal with the (SWB & PWB) & (EWB) restrictions
final_list_flourishing_home = [0] * row;

# Final column professionnal with the (SWB & PWB) & (EWB) restrictions
final_list_languishing_pro = [0] * row;

# Final column personnal with the (SWB & PWB) & (EWB) restrictions
final_list_languishing_home = [0] * row;

# Final column professionnal with the (SWB & PWB) & (EWB) restrictions
final_list_moderate_pro = [0] * row;

# Final column personnal with the (SWB & PWB) & (EWB) restrictions
final_list_moderate_home = [0] * row;

# Final column for Emotional flourishing that measure PE7/NE7
final_emotional_flourishing = [0] * row;
```

```python
# Emotional flourishing measures the ratio between the positive and negative emotions
# It is also called the postivity ratio (PsurN).
# For this ratio, we can use all PE and NE

for index in range(0,row):
    
    positive_row = (D.iloc[index]["PE1"] + D.iloc[index]["PE2"] + D.iloc[index]["PE3"] + D.iloc[index]["PE4"] + D.iloc[index]["PE5"] + D.iloc[index]["PE6"] + D.iloc[index]["PE7"] + D.iloc[index]["PE8"] + D.iloc[index]["PE9"] + D.iloc[index]["PE10"]) / 10
    negative_row = (D.iloc[index]["NE1"] + D.iloc[index]["NE2"] + D.iloc[index]["NE3"] + D.iloc[index]["NE4"] + D.iloc[index]["NE5"] + D.iloc[index]["NE6"] + D.iloc[index]["NE7"] + D.iloc[index]["NE8"] + D.iloc[index]["NE9"] + D.iloc[index]["NE10"]) / 10
    
    if positive_row/negative_row > 3:
        final_emotional_flourishing[index] = 3 # for "emotionally flourishing"
    elif positive_row/negative_row <= 3 and positive_row/negative_row > 2:
        final_emotional_flourishing[index] = 2 # for "moderately emotionally healthy"
    elif positive_row/negative_row <= 2 and positive_row/negative_row >= 1:
        final_emotional_flourishing[index] = 1 # for "emotionally languishing"
    elif positive_row/negative_row < 1:
        final_emotional_flourishing[index] = 0 # for "depressive episode"
    else:
        final_emotional_flourishing[index] = "" # else "empty"

print(final_emotional_flourishing)
```

```python
# Prepare the size matrix for each cell we want to verify
# We use the same row for the repondants
# We use the length of the dimension (nbr of variables)
# to use only those we want to analyse 

find_flourishing_pro = np.zeros((row, len(list_of_flourishing_pro)))
find_flourishing_home = np.zeros((row, len(list_of_flourishing_home)))

find_functioning_questions_home = np.zeros((row, len(list_of_satisfy_functioning_questions_home)))
find_functioning_questions_pro = np.zeros((row, len(list_of_satisfy_functioning_questions_pro)))
```

```python
# Verify one by one the cell 
# And if it's fill Flourishing expectations
# Assign them the value 1 (otherwise 0)

for index in range(0,row):

    for name in list_of_flourishing_pro:
        if D.iloc[index][name] == 5 or D.iloc[index][name] == 6:
            find_flourishing_pro[index][list_of_flourishing_pro.index(name)] = 1
        if D.iloc[index][name] == 3 or D.iloc[index][name] == 4:
            find_flourishing_pro[index][list_of_flourishing_pro.index(name)] = 2
        if D.iloc[index][name] == 1 or D.iloc[index][name] == 2:
            find_flourishing_pro[index][list_of_flourishing_pro.index(name)] = 3
    
    for name in list_of_flourishing_home:
        if D.iloc[index][name] == 5 or D.iloc[index][name] == 6:
            find_flourishing_home[index][list_of_flourishing_home.index(name)] = 1
        if D.iloc[index][name] == 3 or D.iloc[index][name] == 4:
            find_flourishing_home[index][list_of_flourishing_home.index(name)] = 2
        if D.iloc[index][name] == 1 or D.iloc[index][name] == 2:
            find_flourishing_home[index][list_of_flourishing_home.index(name)] = 3

    for name in list_of_positive_functioning_questions_perso:
        if D.iloc[index][name] == 5 or D.iloc[index][name] == 6:
            find_functioning_questions_home[index][list_of_positive_functioning_questions_perso.index(name)] = 1

    for name in list_of_positive_functioning_questions_pro:
        if D.iloc[index][name] == 5 or D.iloc[index][name] == 6:
            find_functioning_questions_pro[index][list_of_positive_functioning_questions_pro.index(name)] = 1
```

```python
# Get the sum of each cell by row to verify the number of 
# positive occurence (if it's superior than 1 or 6...)

sum_of_flourishing_pro_flourishing = (find_flourishing_pro == 1).sum(axis=1)
#print(sum_of_flourishing_pro_flourishing)

sum_of_flourishing_pro_languishing = (find_flourishing_pro == 2).sum(axis=1)
#print(sum_of_flourishing_pro_languishing)

sum_of_flourishing_pro_moderate = (find_flourishing_pro == 3).sum(axis=1)
#print(sum_of_flourishing_pro_moderate)

sum_of_flourishing_home_flourishing = (find_flourishing_home == 1).sum(axis=1)
#print(sum_of_flourishing_home_flourishing)

sum_of_flourishing_home_languishing = (find_flourishing_home == 2).sum(axis=1)
#print(sum_of_flourishing_home_languishing)

sum_of_flourishing_home_moderate = (find_flourishing_home == 3).sum(axis=1)
#print(sum_of_flourishing_home_languishing)

sum_of_functioning_questions_pro = find_functioning_questions_pro.sum(axis=1)
#print(sum_of_functioning_questions_pro)

sum_of_functioning_questions_home = find_functioning_questions_home.sum(axis=1)
#print(sum_of_functioning_questions_home)
```

```python
# Verify for professional context...
# If there is requires a response of "almost every day" (5) or "every day" (6) 
# to 1 or more of the 3 emotional well-being questions (EWB)

# Verify for personnal context...
# If there is requires a response of "almost every day" (5) or "every day" (6) 
# to 1 or more of the 3 emotional well-being questions (EWB)

# For pro, verify for functioning questions (SWB & PWB)....
# If there are 3 or more of the 6 positive functioning questions ! 

# For perso/home, Verify for functioning questions (SWB & PWB)....
# If there are 3 or more of the 6 positive functioning questions ! 

for index in range(0, row):

    if sum_of_flourishing_pro_flourishing[index] >= 1:
        list_of_satisfy_fourishing_pro[index] = 1
    else:
        list_of_satisfy_fourishing_pro[index] = 0

    if sum_of_flourishing_home_flourishing[index] >= 1:
        list_of_satisfy_fourishing_home[index] = 1
    else:
        list_of_satisfy_fourishing_home[index] = 0

    if sum_of_flourishing_pro_languishing[index] >= 1:
        list_of_satisfy_languishing_pro[index] = 1
    else:
        list_of_satisfy_languishing_pro[index] = 0

    if sum_of_flourishing_home_languishing[index] >= 1:
        list_of_satisfy_languishing_home[index] = 1
    else:
        list_of_satisfy_languishing_home[index] = 0

    if sum_of_flourishing_pro_moderate[index] >= 1:
        list_of_satisfy_moderate_pro[index] = 1
    else:
        list_of_satisfy_moderate_pro[index] = 0

    if sum_of_flourishing_home_moderate[index] >= 1:
        list_of_satisfy_moderate_home[index] = 1
    else:
        list_of_satisfy_moderate_home[index] = 0

    if sum_of_functioning_questions_home[index] >= 6:
        list_of_satisfy_functioning_questions_home[index] = 1
    else:
        list_of_satisfy_functioning_questions_home[index] = 0

    if sum_of_functioning_questions_pro[index] >= 6:
        list_of_satisfy_functioning_questions_pro[index] = 1
    else:
        list_of_satisfy_functioning_questions_pro[index] = 0

#print(list_of_satisfy_fourishing_pro)
#print(list_of_satisfy_fourishing_home)

#print(list_of_satisfy_languishing_pro)
#print(list_of_satisfy_languishing_home)

#print(list_of_satisfy_moderate_pro)
#print(list_of_satisfy_moderate_home)

#print(list_of_satisfy_functioning_questions_pro)
#print(list_of_satisfy_functioning_questions_home)
```

```python
# Verify that the 2 conditions are meet to assign the good value "1" (that Flourishing status is good) :
# - The functioning questions (SWB & PWB) rule.
# - The well-being questions (EWB).

for index in range(0, row):
        
    if list_of_satisfy_fourishing_home[index] == 1 and list_of_satisfy_functioning_questions_home[index] == 1:
        final_list_flourishing_home[index] = 1

    if list_of_satisfy_fourishing_pro[index] == 1 and list_of_satisfy_functioning_questions_pro[index] == 1:
        final_list_flourishing_pro[index] = 1

    if list_of_satisfy_languishing_pro[index] == 1 and list_of_satisfy_functioning_questions_pro[index] == 1:
        final_list_languishing_pro[index] = 1

    if list_of_satisfy_languishing_home[index] == 1 and list_of_satisfy_functioning_questions_home[index] == 1:
        final_list_languishing_home[index] = 1

    if (list_of_satisfy_fourishing_pro == 0 or list_of_satisfy_functioning_questions_pro[index] == 0) and (list_of_satisfy_languishing_pro == 0 or list_of_satisfy_functioning_questions_pro[index] == 0):
        final_list_moderate_pro[index] = 1

    if (list_of_satisfy_fourishing_home == 0 or list_of_satisfy_functioning_questions_home[index] == 0) and (list_of_satisfy_languishing_home == 0 or list_of_satisfy_functioning_questions_home[index] == 0):
        final_list_moderate_home[index] = 1
```
