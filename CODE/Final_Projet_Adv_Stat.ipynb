```python
D = pd.read_excel("FINAL_DATASET.xlsx",sheet_name="dataset",index_col=0)

row, col = D.shape

column_names = list(D)
print(column_names)

lst_col = [0] * col

lst_row = [0] * row

iterate_empty_value = 0

iterate_matrix = 0

list_of_deleted_column = []
list_of_deleted_row = []

list_of_flourishing = ['MHCB1', 'MHCC1', 'MHCB2', 'MHCC2', 'MHCB3', 'MHCC3', 'MHCC4', 'MHCC5', 'MHCB6', 'MHCC6', 'MHCB7', 'MHCB8', 'MHCB9', 'MHCC9', 'MHCB10', 'MHCC10', 'MHCB13', 'MHCC13']
list_of_fourishing_pro = ['MHCB1', 'MHCB2', 'MHCB3','MHCB6', 'MHCB7', 'MHCB8', 'MHCB9', 'MHCB10', 'MHCB13']
list_of_flourishing_home =  ['MHCC1', 'MHCC2', 'MHCC3', 'MHCC4', 'MHCC5', 'MHCC6', 'MHCC9', 'MHCC10', 'MHCC13']
list_of_positive_PE = ['PE4','PE7','PE8','PE10']

list_of_satisfy_fourishing_pro = [0] * row;
list_of_satisfy_fourishing_home = [0] * row;

list_of_satisfy_flourishing_low = [0] * row;
list_of_satisfy_flourishing = [0] * row;
list_of_satisfy_positive_pe = [0] * row;

final_list_flourishing = [0] * row;
final_list_languishing = [0] * row;
final_list_moderate_mental = [0] * row;

final_list_flourishing_pro = [0] * row;
final_list_flourishing_home = [0] * row;

final_emotional_flourishing = [0] * row;


for index, ite_row in D.iterrows():
    
    positive_row = ite_row["PE7"]
    negative_row = ite_row["NE7"]
    
    if positive_row/negative_row > 3:
        final_emotional_flourishing[index-1] = "emotionally flourishing"
    elif positive_row/negative_row <= 3 and positive_row/negative_row > 2:
        final_emotional_flourishing[index-1] = "moderately emotionally healthy"
    elif positive_row/negative_row <= 2 and positive_row/negative_row >= 1:
        final_emotional_flourishing[index-1] = "emotionally languishing"
    elif positive_row/negative_row < 1:
        final_emotional_flourishing[index-1] = "depressive episode"
    else:
        final_emotional_flourishing[index-1] = np.NaN


find_flourishing = np.zeros((row, len(list_of_flourishing)))
find_flourishing_low = np.zeros((row, len(list_of_flourishing)))
find_positive_pe = np.zeros((row, len(list_of_positive_PE)))

find_flourishing_pro = np.zeros((row, len(list_of_fourishing_pro)))
find_flourishing_home = np.zeros((row, len(list_of_flourishing_home)))

for index_row, ite_row in D.iterrows():
    
    for name in list_of_flourishing:
        if ite_row[name] == 5 or ite_row[name] == 6 :
            find_flourishing[index_row-1][list_of_flourishing.index(name)] = 1
        if ite_row[name] == 1 or ite_row[name] == 2 :
            find_flourishing_low[index_row-1][list_of_flourishing.index(name)] = 1
            
    for name in list_of_positive_PE:
        if ite_row[name] == 5 or ite_row[name] == 6:
            find_positive_pe[index_row-1][list_of_positive_PE.index(name)] = 1
    
    for name in list_of_fourishing_pro:
        if ite_row[name] == 5 or ite_row[name] == 6:
            find_flourishing_pro[index_row-1][list_of_fourishing_pro.index(name)] = 1
    
    for name in list_of_flourishing_home:
        if ite_row[name] == 5 or ite_row[name] == 6:
            find_flourishing_home[index_row-1][list_of_flourishing_home.index(name)] = 1

#print("List of total number find / sum of 5 and 6 EWB values for flourishing :")            
#print(find_flourishing)

#print("List of total number find / sum of 1 and 2 EWB values for flourishing :")            
#print(find_flourishing_low)

#print("List of total number find / sum of positive PE for flourishing :")   
#print(find_positive_pe)

sum_of_flourishing_low = find_flourishing_low.sum(axis=1)
sum_of_flourishing = find_flourishing.sum(axis=1)
sum_of_positive_pe = find_positive_pe.sum(axis=1)

sum_of_flourishing_pro = find_flourishing_pro.sum(axis=1)
sum_of_flourishing_home = find_flourishing_home.sum(axis=1)

for value in sum_of_flourishing_pro:
    iterate_matrix = iterate_matrix + 1
    if value >= 1:
        list_of_satisfy_fourishing_pro[iterate_matrix-1] = 1
    else:
        list_of_satisfy_fourishing_pro[iterate_matrix-1] = 0

iterate_matrix = 0
    
for value in sum_of_flourishing_home:
    iterate_matrix = iterate_matrix + 1
    if value >= 1:
        list_of_satisfy_fourishing_home[iterate_matrix-1] = 1
    else:
        list_of_satisfy_fourishing_home[iterate_matrix-1] = 0
    
iterate_matrix = 0

for value in sum_of_flourishing_low:
    
    iterate_matrix = iterate_matrix + 1
    
    if value >= 1:
        #print("LANGUISGING - EWB : This row sastisfy the flourishing request low and the ID is : {} ".format(iterate_matrix))
        list_of_satisfy_flourishing_low[iterate_matrix-1] = 1
        
    else:
        #print("LANGUISHING - EWB : This row NOT sastisfy the flourishing request low and the ID is : {} ".format(iterate_matrix))
        list_of_satisfy_flourishing_low[iterate_matrix-1] = 0

#print("LANGUISHING - Final list of once or twice (2) or never (1) EWB : ")
#print(list_of_satisfy_flourishing_low);
    
iterate_matrix = 0;

for value in sum_of_positive_pe:
    
    iterate_matrix = iterate_matrix + 1;
    
    if value >= 1:
        #print("FLOURISHING - Positive PE : This row sastisfy the pe request and the ID is : {} ".format(iterate_matrix))
        list_of_satisfy_positive_pe[iterate_matrix-1] = 1
    else:
        #print("FLOURISHING - Positive PE : This row NOT sastisfy the pe request and the ID is : {} ".format(iterate_matrix))
        list_of_satisfy_positive_pe[iterate_matrix-1] = 0            


#print("LANGUISHING & FLOURISHING - Final list of 6 or more of the 11 positive functioning questions : ")
#print(list_of_satisfy_positive_pe)
    
iterate_matrix = 0;

for value in sum_of_flourishing:
    
    iterate_matrix = iterate_matrix + 1
    
    if value >= 1:
        #print("FLOURISHING - EWB : This row sastisfy the flourishing request and the ID is : {} ".format(iterate_matrix))
        list_of_satisfy_flourishing[iterate_matrix-1] = 1
        
    else:
        #print("FLOURISHING - EWB : This row NOT sastisfy the flourishing request and the ID is : {} ".format(iterate_matrix))
        list_of_satisfy_flourishing[iterate_matrix-1] = 0

#print("FLOURISHING - Final list of almost every day (5) or every day (6) EWB : ")
#print(list_of_satisfy_flourishing);

# final_list_flourishing = [0] * row;
# final_list_languishing = [0] * row;
# final_list_moderate_mental = [0] * row;

iterate_matrix = 0;

for index in range(0, row):
    
    if list_of_satisfy_flourishing[index] == 1 and list_of_satisfy_positive_pe[index] == 1:
        final_list_flourishing[index] = 1
        
    if list_of_satisfy_flourishing_low[index] == 1 and list_of_satisfy_positive_pe[index] == 1:
        final_list_languishing[index] = 1
        
    if list_of_satisfy_fourishing_home[index] == 1 and list_of_satisfy_positive_pe[index] == 1:
        final_list_flourishing_home[index] = 1

    if list_of_satisfy_fourishing_pro[index] == 1 and list_of_satisfy_positive_pe[index] == 1:
        final_list_flourishing_pro[index] = 1
    
print(" ")
print("List of flourishing :")
print(final_list_flourishing)
print(" ")

print("List of languishing : ")
print(final_list_languishing)
print(" ")
        
for index in range(0, row):
    if final_list_flourishing[index] == 0 and final_list_languishing[index] == 0:
        final_list_moderate_mental[index] = 1

print("List of moderate mental : ")
print(final_list_moderate_mental)
print(" ")

print ("List of flourishing - PRO :")
print(final_list_flourishing_pro)
print(" ")

print ("List of flourishing - HOME :")
print(final_list_flourishing_home)
print(" ")

print("List of emotional flourishing")
print(final_emotional_flourishing)
print(" ")

D.assign(Name="EMOTIONAL_FLOURISHING_E7")
D["EMOTIONAL_FLOURISHING_E7"] = final_emotional_flourishing

D.assign(Name="FLOURISHING")
D["FLOURISHING"] = final_list_flourishing

D.assign(Name="LANGUISHING")
D["LANGUISHING"] = final_list_languishing

D.assign(Name="MODERATE_MENTAL")
D["MODERATE_MENTAL"] = final_list_moderate_mental

D.assign(Name="FLOURISHING_PRO")
D["FLOURISHING_PRO"] = final_list_flourishing_pro

D.assign(Name="FLOURISHING_HOME")
D["FLOURISHING_HOME"] = final_list_flourishing_home

D.head(100)
        
```
