PART 1 – Data Cleaning an Preparation

```python
# Read excel data...
D = pd.read_excel("FINAL_DATASET.xlsx",sheet_name="dataset",index_col=0)

# Assign to row and column the size of instances & variables
row, col = D.shape

# Get all variables names in a list
column_names = list(D)

# Create a list of 0 x Number of variables size. For verifying after the empty values...
# With the value "7" don't know...
lst_col_with_no_know = [0] * col

# Create a list of 0 x Number of instances size. For verifying after the empty values...
# With the value "7" don't know...
lst_row_with_no_know = [0] * row

# Create a list of 0 x Number of variables size. For verifying after the empty values...
lst_col = [0] * col

# Create a list of 0 x Number of instances size. For verifying after the empty values...
lst_row = [0] * row

# Help to iterate on the row to see how many empty value remain for each respondant
iterate_empty_value = 0

# Help to iterate on the row to see how many empty value remain for each respondant...
# Integrate "7" values (don't know)
iterate_empty_value_with_no_know = 0

# Help to iterate inside a matrix and get the respective index to assign to other list
iterate_matrix = 0

# The list of column names that didn't respect the minimum of 50% value available per variable
# With "don't know" value
list_of_deleted_column_with_no_know = []

# The list of row id that didn't respect the minimum of 50% value available per variable
# With "don't know" value
list_of_deleted_row_with_no_know = []

# The list of column names that didn't respect the minimum of 50% value available per variable
list_of_deleted_column = []

# The list of row id that didn't respect the minimum of 50% value available per instance
list_of_deleted_row = []

# The list of variables for professionnal and well-being questions (EWB)
list_of_flourishing_pro = ['MHCB1', 'MHCB2', 'MHCB3'] #,'MHCB6', 'MHCB7', 'MHCB8', 'MHCB9', 'MHCB10', 'MHCB13']

# The list of variables for personnal and well-being questions (EWB)
list_of_flourishing_home =  ['MHCC1', 'MHCC2', 'MHCC3'] #, 'MHCC4', 'MHCC5', 'MHCC6', 'MHCC9', 'MHCC10', 'MHCC13']

# The list of variables for professionnal and (SWB & PWB)
list_of_positive_functioning_questions_pro = ['MHCB6', 'MHCB7', 'MHCB8', 'MHCB9', 'MHCB10', 'MHCB13']

# The list of variables for personnal and (SWB & PWB)
list_of_positive_functioning_questions_perso = ['MHCC4', 'MHCC5', 'MHCC6', 'MHCC9', 'MHCC10', 'MHCC13']

# For PRO : List that verify according to the sum of validate value if well-being questions is respected
list_of_satisfy_fourishing_pro = [0] * row;

# For HOME : List that verify according to the sum of validate value if well-being questions is respected
list_of_satisfy_fourishing_home = [0] * row;

# For PRO : List that verify according to the sum of validate value if positive functioning questions is respected
list_of_satisfy_functioning_questions_pro = [0] * row;

# For HOME : List that verify according to the sum of validate value if positive functioning questions is respected
list_of_satisfy_functioning_questions_home = [0] * row;

# Final column professionnal with the (SWB & PWB) & (EWB) restrictions
final_list_flourishing_pro = [0] * row;

# Final column personnal with the (SWB & PWB) & (EWB) restrictions
final_list_flourishing_home = [0] * row;

# Final column for Emotional flourishing that measure PE7/NE7
final_emotional_flourishing = [0] * row;

# Percentage of empty values dedicated to the row/instances
percentage_empty_row_list = [0] * row;

# Percentage of empty values dedicated to the row/instances (with "don't know" value)
percentage_empty_row_list_with_no_know  = [0] * row;
```

```python
# Emotional flourishing measures the ratio between the positive and negative emotions
# It is also called the postivity ratio (PsurN).
# For this ratio, we can only use PE7 and NE7
# We don't have the rest of the variables...

for index, ite_row in D.iterrows():
    
    positive_row = ite_row["PE7"]
    negative_row = ite_row["NE7"]
    
    if positive_row/negative_row > 3:
        final_emotional_flourishing[index-1] = "emotionally flourishing"
    elif positive_row/negative_row <= 3 and positive_row/negative_row > 2:
        final_emotional_flourishing[index-1] = "moderately emotionally healthy"
    elif positive_row/negative_row <= 2 and positive_row/negative_row >= 1:
        final_emotional_flourishing[index-1] = "emotionally languishing"
    elif positive_row/negative_row < 1:
        final_emotional_flourishing[index-1] = "depressive episode"
    else:
        final_emotional_flourishing[index-1] = 0
```

```python
# Prepare the size matrix for each cell we want to verify
# We use the same row for the repondants
# We use the length of the dimension (nbr of variables)
# to use only those we want to analyse 

find_flourishing_pro = np.zeros((row, len(list_of_flourishing_pro)))
find_flourishing_home = np.zeros((row, len(list_of_flourishing_home)))

find_functioning_questions_home = np.zeros((row, len(list_of_satisfy_functioning_questions_home)))
find_functioning_questions_pro = np.zeros((row, len(list_of_satisfy_functioning_questions_pro)))
```

```python
# Verify one by one the cell 
# And if it's fill Flourishing expectations
# Assign them the value 1 (otherwise 0)

for index_row, ite_row in D.iterrows():
    
    for name in list_of_flourishing_pro:
        if ite_row[name] == 5 or ite_row[name] == 6:
            find_flourishing_pro[index_row-1][list_of_flourishing_pro.index(name)] = 1
    
    for name in list_of_flourishing_home:
        if ite_row[name] == 5 or ite_row[name] == 6:
            find_flourishing_home[index_row-1][list_of_flourishing_home.index(name)] = 1

    for name in list_of_positive_functioning_questions_perso:
        if ite_row[name] == 5 or ite_row[name] == 6:
            find_functioning_questions_home[index_row-1][list_of_positive_functioning_questions_perso.index(name)] = 1

    for name in list_of_positive_functioning_questions_pro:
        if ite_row[name] == 5 or ite_row[name] == 6:
            find_functioning_questions_pro[index_row-1][list_of_positive_functioning_questions_pro.index(name)] = 1
```

```python
# Get the sum of each cell by row to verify the number of 
# positive occurence (if it's superior than 1 or 6...)

sum_of_flourishing_pro = find_flourishing_pro.sum(axis=1)
sum_of_flourishing_home = find_flourishing_home.sum(axis=1)

sum_of_functioning_questions_pro = find_functioning_questions_pro.sum(axis=1)
sum_of_functioning_questions_home = find_functioning_questions_home.sum(axis=1)
```

```python
# Verify for professional context...
# If there is requires a response of "almost every day" (5) or "every day" (6) 
# to 1 or more of the 3 emotional well-being questions (EWB)

for value in sum_of_flourishing_pro:
    iterate_matrix = iterate_matrix + 1
    if value >= 1:
        list_of_satisfy_fourishing_pro[iterate_matrix-1] = 1
    else:
        list_of_satisfy_fourishing_pro[iterate_matrix-1] = 0
```

```python
# Verify for personnal context...
# If there is requires a response of "almost every day" (5) or "every day" (6) 
# to 1 or more of the 3 emotional well-being questions (EWB)

iterate_matrix = 0
    
for value in sum_of_flourishing_home:
    iterate_matrix = iterate_matrix + 1
    if value >= 1:
        list_of_satisfy_fourishing_home[iterate_matrix-1] = 1
    else:
        list_of_satisfy_fourishing_home[iterate_matrix-1] = 0
```

```python
# For pro, verify for functioning questions (SWB & PWB)....
# If there are 3 or more of the 6 positive functioning questions ! 

iterate_matrix = 0
    
for value in sum_of_functioning_questions_pro:
    iterate_matrix = iterate_matrix + 1
    if value >= 3:
        list_of_satisfy_functioning_questions_pro[iterate_matrix-1] = 1
    else:
        list_of_satisfy_functioning_questions_pro[iterate_matrix-1] = 0
```

```python
# For perso/home, Verify for functioning questions (SWB & PWB)....
# If there are 3 or more of the 6 positive functioning questions ! 

iterate_matrix = 0
    
for value in sum_of_functioning_questions_home:
    iterate_matrix = iterate_matrix + 1
    if value >= 3:
        list_of_satisfy_functioning_questions_home[iterate_matrix-1] = 1
    else:
        list_of_satisfy_functioning_questions_home[iterate_matrix-1] = 0
```

```python
# Verify that the 2 conditions are meet to assign the good value "1" (that Flourishing status is good) :
# - The functioning questions (SWB & PWB) rule.
# - The well-being questions (EWB).

iterate_matrix = 0;

for index in range(0, row):
        
    if list_of_satisfy_fourishing_home[index] == 1 and list_of_satisfy_functioning_questions_home[index] == 1:
        final_list_flourishing_home[index] = 1

    if list_of_satisfy_fourishing_pro[index] == 1 and list_of_satisfy_functioning_questions_pro[index] == 1:
        final_list_flourishing_pro[index] = 1
```

```python
# Print final column values for professionnal (cell = 1 is Flourishing status OK, otherwise it's 0)
print ("List of flourishing - PRO :")
print(final_list_flourishing_pro)
print(" ")

# Print final column values for personnal (cell = 1 is Flourishing status OK, otherwise it's 0)
print ("List of flourishing - HOME :")
print(final_list_flourishing_home)
print(" ")

# Print categorical values of Emotional flourishing according PE7/NE7 columns.
print("List of emotional flourishing")
print(final_emotional_flourishing)
print(" ")
```

```python
# Add new column EMOTIONAL_FLOURISHING_E7 in the dataset
D.assign(Name="EMOTIONAL_FLOURISHING_E7")
D["EMOTIONAL_FLOURISHING_E7"] = final_emotional_flourishing

# Add new column FLOURISHING_PRO in the dataset
D.assign(Name="FLOURISHING_PRO")
D["FLOURISHING_PRO"] = final_list_flourishing_pro

# Add new column FLOURISHING_HOME in the dataset
D.assign(Name="FLOURISHING_HOME")
D["FLOURISHING_HOME"] = final_list_flourishing_home
```

```python
# FOR ROWS
list_of_deleted_row = [];

print("Test for rows ...")
for index, ite_row in D.iterrows():
    
    iterate_empty_value = 0

    iterate_empty_value_with_no_know = 0
    
    print(" ")
    print("Row number id dataframe : {} ".format(index))
    print("Row GUI : {} ".format(ite_row["GUI"]))
    
    for ite_column in column_names:
        
        my_row = ite_row[ite_column]
        
        #print("For cell column : {} ".format(ite_column))
        #print("Value is : {} ".format(my_row))
        
        if  my_row == "" or my_row != my_row :
            iterate_empty_value = iterate_empty_value + 1

        if my_row == "" or my_row != my_row or my_row == "7" or my_row == 7 :
            iterate_empty_value_with_no_know = iterate_empty_value_with_no_know + 1

        
            
    lst_row[index-1] = iterate_empty_value
    lst_row_with_no_know[index-1] = iterate_empty_value_with_no_know
        
    print("Number of empty values : {} ".format(lst_row[index-1]))
        
    percentage = (lst_row[index-1] / col) * 100
    percentage_with_no_know = (lst_row_with_no_know[index-1] / col) * 100
        
    print("Percentage of empty values : {} ".format(percentage))
    percentage_empty_row_list.append(percentage)

    print("Percentage of empty values (with don't know value) : {}".format(percentage_with_no_know))
    percentage_empty_row_list_with_no_know.append(percentage_with_no_know)
        
    if percentage > 50:
        print("Warning ROW ! Percentage of empty data > 50% for index : {} ".format(index))
        list_of_deleted_row.append(index)

    if percentage_with_no_know > 50:
        list_of_deleted_row_with_no_know.append(index)
```

```python
# FOR COLUMNS
list_of_deleted_column = [];

print("Test for columns ...")
for item in column_names:
    
    iterate_empty_value = 0

    iterate_empty_value_with_no_know = 0
    
    print(" ")
    print("Column number id dataframe : {} ".format(list(D.columns).index(item)))
    print("Column name : {} ".format(column_names[list(D.columns).index(item)]))
    
    for info in D[item]:
        
        if info == "" or info != info :
            iterate_empty_value = iterate_empty_value + 1;

        if info == "" or info != info or info == "7" or info == 7 :
            iterate_empty_value_with_no_know = iterate_empty_value_with_no_know + 1
            
    lst_col[list(D.columns).index(item)] = iterate_empty_value
    lst_col_with_no_know[list(D.columns).index(item)] = iterate_empty_value_with_no_know
    
        #else:
            #lst_col[list(D.columns).index(item)] = "Good"
    print("Number of empty values : {} ".format(lst_col[list(D.columns).index(item)]))
    
    percentage = (lst_col[list(D.columns).index(item)] / row) * 100
    percentage_with_no_know = (lst_col_with_no_know[list(D.columns).index(item)] / row) * 100
    
    print("Percentage of empty values : {} ".format(percentage))
    print("Percentage of empty values (with don't know value) : {}".format(percentage_with_no_know))
    
    if percentage > 50:
        print("Warning COL ! Percentage of empty data > 50% for : {} ".format(column_names[list(D.columns).index(item)]))
        list_of_deleted_column.append(column_names[list(D.columns).index(item)])

    if percentage_with_no_know > 50:
        list_of_deleted_column_with_no_know.append(column_names[list(D.columns).index(item)])
```

```python
print("Delete unusable data empty by < 50% ")
print(" ")

# Delete the row with 50% of empty values...
print("List of rows to delete : {} ".format(list_of_deleted_row))
print(" ")

# Delete the columns with 50% of empty values...
print("List of columns to delete : {} ".format(list_of_deleted_column))
print(" ")

# Delete the rows with 50% of empty value AND considering that "don't know value" is a empty value also (value that is equal to 7)
print("List of rows to delete with 'don't know' values : {} ".format(list_of_deleted_row_with_no_know))
print(" ")

# Delete the columns with 50% of empty value AND considering that "don't know value" is a empty value also (value that is equal to 7)
print("List of columns to delete with 'don't know' values : {} ".format(list_of_deleted_column_with_no_know))
print(" ")

difference_id = list((set(list_of_deleted_row_with_no_know)- set(list_of_deleted_row))) + list((set(list_of_deleted_row)- set(list_of_deleted_row_with_no_know)))
print("The difference of data between the two row lists (with and without don't know values) are : ",difference_id)
print(" ")

print("List by row of percentage of empty value: ",percentage_empty_row_list)
print(" ")

print("List by row of percentage of empty value (with 'don't know' values) : ",percentage_empty_row_list_with_no_know)
print(" ")


D.drop(list_of_deleted_column, axis=1, inplace=True);
D.drop(index=list_of_deleted_row, inplace = True);
```
